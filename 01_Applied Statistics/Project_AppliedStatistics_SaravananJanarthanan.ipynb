
**Question 2**
An electrical manufacturing company conducts quality checks at specified periods on the products it manufactures. Historically, the failure rate for the manufactured item is 5%. Suppose a random sample of 10 manufactured items is selected. 

**Q 2.A**
 Probability that none of the items are defective?

**Ans**
Binomial distribution is used in quality control and quality control and quality assurance hence can be used for above defective analysis
Since 10 items are selected, the defective number probability varies  from 0 to 10 , 11 possibilities or quantiles (**samp_rng**)
using python binomal function probability mass function to ascertain the distribution by passing all possible values of defective rate **def_rate** i.e 5% or 0.05 for a range 0 to sample number **samp_num** i.e 10.
None of them defective lies in the 1st  position of the distribution or 0 index

```python
# defective or failure rate
def_rate = 0.05 

# sample numbers
samp_num = 10 # 

# possible values or array of quantiles of definite 
samp_rng = np.arange( 0,11 )

#binomial distribution of possible defective rate 
bin_posb_val = stats.binom.pmf( samp_rng, samp_num, def_rate )

print( color.BOLD +'Answer 2B '+color.END+': Probability that none of the items are defective is {:.2f}% ({:.5f})'.format( bin_posb_val[0]*100, bin_posb_val[0] ) )
```

**Q 2.B**
Probability that exactly one of the items is defective?</br>

**Ans**
Probablity of exactly one of the items is defective is lies in the second position of the binomial distribution or 1 of the index

```python
print( color.BOLD +'Answer 2B'+color.END+': Probability that exactly one of the items defective is {:.2f}% ({:.5f})'.format( bin_posb_val[1]*100, bin_posb_val[1] ) )
```


**Q 2.C.** Probability that two or fewer of the items are defective?

**Ans** The probability of two or lesser items being defective is the cumulative of the defective probability distribution from  0 to 2. 
Using the 	Cumulative density function the cumulative distribution for 0 to 2 can be arrived or use the sum of the probabilities of 0, 1 & 2 from probability mass function.

```python
none_deft = bin_posb_val[0]+ bin_posb_val[1] + bin_posb_val[2]
print( color.BOLD +'Answer 2C'+color.END+': Probability that none of the items are defective is {:.2f}% ({:.5f})'.format( none_deft*100, none_deft ) )
```

```python
#Cumulative density function
cum_den_val = stats.binom.cdf( samp_rng, samp_num, def_rate )
print( color.BOLD +'Answer 2C'+color.END+': Probability that none of the items are defective is {:.2f}% ({:.5f})'.format( cum_den_val[2]*100, cum_den_val[2] ) )
```


**Q 2.D.** Probability that three or more of the items are defective ?

**Ans.** Simillar to 2C solution the possible values form three or more items being defecive is sum of the probabilites from 3 to 10 or
deducting the cumulative probability of upto 2 items from max cumulative probability or
deducting the cumulative probability of upto 2 items from  1 which is the sum of the all the probabilities from the possible probabaility distribution

```python
# sum of the probabilites from 3 to 10
sum_3_to_10 = bin_posb_val[3]+ bin_posb_val[4] + bin_posb_val[5]+bin_posb_val[6]+ bin_posb_val[7] + bin_posb_val[8]+bin_posb_val[9]+ bin_posb_val[10]
print( color.BOLD +'Answer 2D'+color.END+': Probability that none of the items are defective is {:.2f}% ({:.5f})'.format( sum_3_to_10*100, sum_3_to_10 ) )
```

```python
#deducting the cumulative probability of upto 2 items from max cumulative probability
cum_prob1 = cum_den_val[10] - cum_den_val[2]
print( color.BOLD +'Answer 2D'+color.END+': Probability that none of the items are defective is {:.2f}% ({:.5f})'.format( cum_prob1*100, cum_prob1 ) )
```

```python
#deducting the cumulative probability of upto 2 items from 1 i.e from sum of the all the probabilities
cum_prob2 = 1 - cum_den_val[2]
print( color.BOLD +'Answer 2D'+color.END+': Probability that none of the items are defective is {:.2f}% ({:.5f})'.format( cum_prob2*100, cum_prob2 ) )
```


**Question 3** 
A car salesman sells on an average 3 cars per week</br>

**Q 3.A.**
What is Probability that in a given week he will sell some cars?

**Ans.** 
The mean or average the car salesman sells cars is 3 per week
to identify the probability of car sales for a given week that is a fixed interval of time , without fixed number of distributions use poisson distribution

mean or average (λ) is 3 (**pois_lambda**)

max numbers of cars the salesman could have sold is assumed as 10 (in a week might have sold nothing upto 10  )
Possibility ranges from 0 to 10 (**sale_range**)

The probability that salesman will sell some cars is from 1 car sales and above, i.e summing up all the probabilities from 1 to max car sales 

```python
# mean or average (λ) is 3
pois_lambda  = 3

#max numbers of cars the salesman could have sold 
max_car_sale = 10

#Possibility ranges
sale_range   = np.arange( 0, max_car_sale+1 )

#poisson distribution
cuml_poisson_dist = stats.poisson.cdf( sale_range , pois_lambda )

#The probability that salesman will sell some cars in a given week 1- cumulative of none i.e 1st distribution of probability
cum_poiss_some = 1 - cuml_poisson_dist[0]
print(color.BOLD +'Answer 3A '+color.END+': Probability that salesman will sell some cars in a given week is {:.2f}% ({:.5f})'.format( cum_poiss_some*100, cum_poiss_some ) )

```


**3.B.** 
What is Probability that in a given week he will sell 2 or more but less than 5 cars?

**Ans.** The Probability that in a given week he will sell 2 or more but less than 5 cars is summing up the probabilities selling 2 cars, 3 cars and 4 cars a week


```python
# Use poisson probability mass function
poiss_pmf = stats.poisson.pmf( sale_range , pois_lambda )

car_sale_2_to_4 = poiss_pmf[2]+ poiss_pmf[3]+poiss_pmf[4]
print( color.BOLD +'Answer 3B '+color.END+': Probability that in a given week he will sell 2 or more but less than 5 cars is {:.2f}% ({:.4f}) '.format( car_sale_2_to_4*100,  car_sale_2_to_4 ))
```


**Question 4**

Accuracy in understanding orders for a speech based bot at a restaurant is important for the Company X which has designed,
marketed and launched the product for a contactless delivery due to the COVID-19 pandemic. Recognition accuracy that measures the percentage of orders that are taken correctly is 86.8%. Suppose that you place an order with the bot and two friends of yours independently place orders with the same bot. Answer the following questions.

**Q 4.A.**

What is the probability that all three orders will be recognised correctly?

Since the outcome has two possibilities either all 3 orders are recognized by the bot or not ,  using binomial distribution identify the probabiity of success of 3 orders going through with a sucess rate of 86.8%
all three orders going through is the 4th outcome out of total 4 outcomes

1st outcome - none of the three orders getting through

2nd outcome - 1 out of 3 orders getting through

3rd outcome - 2 out of 3 orders getting through

4th outcome - 3 out of 3 orders getting through

```python
# Sucess rate
sux_rt = 0.868

# number of trials 
trial_num = 3

# possible outcomes
pos_outcm = np.arange( 0,trial_num+1 )

binom_sucss = stats.binom.pmf( pos_outcm, trial_num, sux_rt )

print( color.BOLD +'Answer 4A'+color.END+': Probability that all three orders will be recognised correctly is {:.2f}% ({:.5f}) '.format( binom_sucss[3]*100,  binom_sucss[3] ) ) 
```


**Q 4.B.**
What is the probability that none of the three orders will be recognised correctly?


**Ans** The probability that none of the three orders will be recognised correctly is 1st outcome - none of the three orders getting through

```python
print( color.BOLD +'Answer 4B'+color.END+': Probability that none of the three orders will be recognised correctly is {:.2f}% ({:.5f}) '.format( binom_sucss[0]*100,  binom_sucss[0] ) ) 
```


**Q 4.C.** What is the probability that at least two of the three orders will be recognised correctly?

**Ans.** The probability that at least two of the three orders will be recognised correctly is sum of 3rd outcome and 4th outcome, the success ranging between 2 to 3 orders going through

```python
# sum of probabilities of 2 order and 3 order placed sucessfully
prob_1_to_2_sucss = binom_sucss[2] + binom_sucss[3]
print( color.BOLD +'Answer 4B'+color.END+': Probability that at least two of the three orders will be recognised correctly {:.2f}% ({:.5f}) '.format( prob_1_to_2_sucss*100,  prob_1_to_2_sucss ) )
```

**Q 4B.** Write the null hypothesis and alternative hypothesis. 

**Ans 4B**

**Null Hypothesis (Ho)**      :  There is no difference between funds sourced by operating and closed companies</br>
**Alternate Hypothesis (Ha)** :  There is a significant difference between funds sourced by operating and closed companies

**Q 4C.** Test for significance and conclusion

```python
# import Z test and t test libraries
from statsmodels.stats.weightstats import ztest
from scipy.stats import ttest_ind
```

```python

# Create samples from operating companies and from closed companies
opertng_sample = part_c_data_wo_outlier.loc[ part_c_data_wo_outlier.OperatingState == 'Operating', 'Funds_in_million' ]
clsd_sample    = part_c_data_wo_outlier.loc[ part_c_data_wo_outlier.OperatingState == 'Closed', 'Funds_in_million' ]

#consider a 0.05 or 5% significance levels to validate the hypothesis
alpha_consrd = 0.05

#Execute the Z test using 2 samples created
ztest_statistics, z_p_value = ztest( opertng_sample, clsd_sample )

if z_p_value <= alpha_consrd :
    print(f'As the p-value from ztest, { round( z_p_value, 3 ) } <  { alpha_consrd } alpha value, there is significant difference between funds sourced by operating & closed  companies, hence the Null Hypothesis is rejected')
else :
    print(f'As the p-value from ztest, { round( z_p_value, 3 ) } >  { alpha_consrd } alpha value, there is NO significant difference between funds sourced by operating & closed  companies, hence FAIL to reject the Null Hypothesis')
```

```python
#Execute the t test using 2 samples created
ttest_statistics, t_p_value = ttest_ind( opertng_sample, clsd_sample )

if t_p_value <= alpha_consrd :
    print(f'As the p-value from ztest, { round( t_p_value, 3 ) } <  { alpha_consrd } alpha value, there is significant difference between funds sourced by operating & closed  companies, hence the Null Hypothesis is rejected')
else :
    print(f'As the p-value from ztest, { round( t_p_value, 3 ) } >  { alpha_consrd } alpha value, there is NO significant difference between funds sourced by operating & closed  companies, hence FAIL to reject the Null Hypothesis')
```

**Answer 4C.** From the Z test and t test it is concluded that there is NO significant difference between funds sourced by operating & closed  companies, hence **FAIL to reject the Null Hypothesis**
